```python
import cv2
from cvzone.HandTrackingModule import HandDetector
import pyautogui as pyg
```

```python
cap = cv2.VideoCapture(0)
detector = HandDetector()
while True:
    status, photo = cap.read()
    hands, img = detector.findHands(photo,flipType=True)
    if hands:
        # Information for the first hand detected
        hand1 = hands[0]  # Get the first hand detected
        lmList1 = hand1["lmList"]  # List of 21 landmarks for the first hand
        handType1 = hand1["type"]  # Type of the first hand ("Left" or "Right")

        # Count the number of fingers up for the first hand
        Upfingers = detector.fingersUp(hand1)
#         print(fingers1)
        if Upfingers.count(1) == 5:
            pyg.press('s')
        elif Upfingers[0] == 1 & Upfingers.count(1) == 1:
            pyg.press('a')
        elif Upfingers[4] == 1 & Upfingers.count(1) == 1:
            pyg.press('d')
        elif Upfingers[1] == 1 & Upfingers.count(1) == 1:
            pyg.press('w')
    cv2.imshow("video-stream",img)
    if cv2.waitKey(50) == 13:
        break
cap.release()
cv2.destroyAllWindows()
```
