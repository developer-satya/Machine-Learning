```python
import cv2
import mediapipe as mp
import math
import pyautogui
import screen_brightness_control as sbc
 


# Function to calculate distance between two points
def calculate_distance(point1, point2):
    return math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

# Function to detect hand gesture

def customize_volume(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.THUMB_TIP]
    index_finger_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.INDEX_FINGER_TIP]
    
    thumb_tip_x, thumb_tip_y = thumb_tip.x, thumb_tip.y
    index_finger_tip_x, index_finger_tip_y = index_finger_tip.x, index_finger_tip.y
    
    # Calculate distance between thumb tip and index finger tip
    distance = calculate_distance((thumb_tip_x, thumb_tip_y), (index_finger_tip_x, index_finger_tip_y))
    # Adjust this threshold based on your hand's size and the distance between thumb and index finger
    if distance < 0.1:
        pyautogui.press('volumedown')
    elif distance > 0.5:
        pyautogui.press('volumeup')  
            
def customize_brightness(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.THUMB_TIP]
    index_finger_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.INDEX_FINGER_TIP]
    
    thumb_tip_x, thumb_tip_y = thumb_tip.x, thumb_tip.y
    index_finger_tip_x, index_finger_tip_y = index_finger_tip.x, index_finger_tip.y
    
    # Calculate distance between thumb tip and index finger tip
    distance = calculate_distance((thumb_tip_x, thumb_tip_y), (index_finger_tip_x, index_finger_tip_y))
    # Adjust this threshold based on your hand's size and the distance between thumb and index finger
    if distance < 0.1:
        sbc.fade_brightness(finish=100, increment = 5)
    elif distance > 0.5:
        sbc.fade_brightness(finish=0, increment = -5)
            
        

# Main function to capture video feed and detect gestures
def main():
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands()
    cap = cv2.VideoCapture(0)
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Convert the image to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Detect hand landmarks
        results = hands.process(frame_rgb)
        
        # Check if hand landmarks are detected
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                if hand_landmarks == results.multi_hand_landmarks[0]:
                    customize_volume(hand_landmarks)
                else:
                    customize_brightness(hand_landmarks)
                
        
        cv2.imshow('Hand Tracking', frame)
        if cv2.waitKey(1)& 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import mediapipe as mp
import math
import pyautogui
import screen_brightness_control as sbc

# Function to calculate distance between two points
def calculate_distance(point1, point2):
    return math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

# Function to detect hand gesture
def detect_gesture(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.THUMB_TIP]
    index_finger_tip = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.INDEX_FINGER_TIP]
    
    thumb_tip_x, thumb_tip_y = thumb_tip.x, thumb_tip.y
    index_finger_tip_x, index_finger_tip_y = index_finger_tip.x, index_finger_tip.y
    
    # Calculate distance between thumb tip and index finger tip
    distance = calculate_distance((thumb_tip_x, thumb_tip_y), (index_finger_tip_x, index_finger_tip_y))
    
    # Adjust this threshold based on your hand's size and the distance between thumb and index finger
    if distance < 0.1:
        return "increase_volume"
    elif distance > 0.5:
        return 'decrease_volume'
    else:
        return None

# Main function to capture video feed and detect gestures
def main():
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands()
    cap = cv2.VideoCapture(0)
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        # Convert the image to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Detect hand landmarks
        results = hands.process(frame_rgb)
        
        # Check if hand landmarks are detected
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                if hand_landmarks == results.multi_hand_landmarks[0]:  # Assuming first hand is left hand
                    # Adjust brightness using left hand
                    THUMB_TIP = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.THUMB_TIP]
                    INDEX_TIP = hand_landmarks.landmark[mp.solutions.hands.HandLandmark.INDEX_FINGER_TIP]
                    distance = calculate_distance((THUMB_TIP.x, THUMB_TIP.y), (INDEX_TIP.x, INDEX_TIP.y))
                    distance = int(distance * 255)
                    print(distance)
                    sbc.set_brightness(distance%100)
                    
                else:
                    # Adjust volume using right hand
                    gesture = detect_gesture(hand_landmarks)
                    if gesture == "increase_volume":
                        pyautogui.press('volumeup')  # Adjust volume up
                    elif gesture == 'decrease_volume':
                        pyautogui.press('volumedown')
        
        cv2.imshow('Hand Tracking', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
