```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150,
                 n_features=2,
                 centers=3,
                 random_state=0)
X
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:,1], c="blue", marker="o", s=50)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3,
            init="k-means++",
           n_init=100,
           max_iter=1000)
y_km = km.fit_predict(X)
```

```python
y_km
```

```python
plt.scatter(X[y_km==0,0],
           X[y_km==0,1],
           s=50,
           c='green',
           marker='s',
           label='cluster 1')
plt.scatter(X[y_km==1,0],
           X[y_km==1,1],
           s=50,
           c='blue',
           marker='o',
           label='cluster 2')
plt.scatter(X[y_km==2,0],
           X[y_km==2,1],
           s=50,
           c='red',
           marker='*',
           label='cluster 3')

plt.scatter(X[y_km==0,0],
           X[y_km==0,1],
           s=50,
           c='green',
           marker='s',
           label='cluster 1')
```

```python
km.inertia_
```

### Using the elbow method to find number of cluster

```python
distortion = []
for i in range(1,8):
    km = KMeans(n_clusters=i,
            init="k-means++",
           n_init=10,
           max_iter=300)
    km.fit(X)
    distortion.append(km.inertia_)
plt.plot(range(1,8),distortion, marker='o',)
plt.xlabel("Number of Clusters")
plt.ylabel("Distortion")
plt.show()
```
